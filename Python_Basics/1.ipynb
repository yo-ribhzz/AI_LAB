```python
# Sum of digits of a number 
num = int(input("Enter a number: "))
sum_digits = 0

while num > 0:
    sum_digits += num % 10
    num //= 10

print("Sum of digits =", sum_digits)
```

```python
# Count number of digits 
num = int(input("Enter a number: "))
count = 0

if num == 0:
    count = 1
else:
    while num > 0:
        num //= 10
        count += 1

print("Number of digits:", count)
```

```python
# Print Prime numbers from 1 to 50 
print("Prime numbers from 1 to 50 are:")
for num in range(2, 51):
    isPrime = True
    for i in range(2, num):
        if num % i == 0:
            isPrime = False
            break
    if isPrime:
        print(num, end=" ")
```

```python
# Implement Stack 
stack = []  

def push(x):
    stack.append(x)
    print(x, "pushed")

def pop():
    if len(stack) == 0:
        print("Stack Underflow")
    else:
        print(stack.pop(), "popped")

def display_stack():
    if len(stack) == 0:
        print("Stack is Empty")
    else:
        print("Stack (top → bottom):", stack[::-1])

push(10)
push(20)
push(30)
display_stack()
pop()
display_stack()
```

```python
# Implement Queue 
queue = []  

def enqueue(x):
    queue.append(x)
    print(x, "enqueued")

def dequeue():
    if len(queue) == 0:
        print("Queue Underflow")
    else:
        print(queue.pop(0), "dequeued")

def display_queue():
    if len(queue) == 0:
        print("Queue is Empty")
    else:
        print("Queue (front → rear):", queue)

enqueue(5)
enqueue(10)
enqueue(15)
display_queue()
dequeue()
display_queue()
```

```python
# Adjacency Matrix 
def build_graph_simple():

    # Step 1: take number of nodes
    n = int(input("How many nodes? "))     # Example: 5 nodes → 1,2,3,4,5

    # Step 2: create empty adjacency list
    adj = [[] for _ in range(n + 1)]

    # Step 3: take edges one by one
    m = int(input("How many edges? "))     # Example: enter 4 edges

    print("\nEnter edges like: 1 2 (means connection between node 1 and 2)")
    for i in range(m):
        u, v = map(int, input(f"Edge {i+1}: ").split())
        adj[u].append(v)
        adj[v].append(u)    # remove this line if graph is directed (one-way)

    # Step 4: print adjacency list clearly
    print("\nAdjacency List:")
    for i in range(1, n + 1):
        print(i, "->", adj[i])


# Run program
build_graph_simple()
```

```python
# 5 X 5 Grid 
def grid_5x5():
    grid = [["." for _ in range(5)] for _ in range(5)]

    start = (0, 0)
    goal = (4, 4)

    grid[start[0]][start[1]] = "S"
    grid[goal[0]][goal[1]] = "G"

    print("\n5x5 Grid:")
    for row in grid:
        print(" ".join(row))


grid_5x5()
```
