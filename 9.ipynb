```python
import heapq

ROWS, COLS = 5, 5

start = (0, 0)
treasure = (4, 4)

def manhattan_distance(current, goal):
    return abs(current[0] - goal[0]) + abs(current[1] - goal[1])

def best_first_search(start, treasure):
    open_list = []
    heapq.heappush(open_list, (manhattan_distance(start, treasure), start))

    visited = set()
    parent = {}

    while open_list:
        _, current = heapq.heappop(open_list)

        if current in visited:
            continue

        visited.add(current)

        if current == treasure:
            print("Treasure found!")
            return reconstruct_path(parent, start, treasure)

        x, y = current

        directions = [(-1,0), (1,0), (0,-1), (0,1)]

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            neighbor = (nx, ny)

            if 0 <= nx < ROWS and 0 <= ny < COLS and neighbor not in visited:
                parent[neighbor] = current
                h = manhattan_distance(neighbor, treasure)
                heapq.heappush(open_list, (h, neighbor))

    print("Treasure not found")
    return None

def reconstruct_path(parent, start, treasure):
    path = [treasure]
    while path[-1] != start:
        path.append(parent[path[-1]])
    path.reverse()
    return path

path = best_first_search(start, treasure)

print("Path taken:")
for step in path:
    print(step)
```
