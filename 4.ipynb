```python
# Q5: Plot various graphs using matplotlib

import numpy as np
import matplotlib.pyplot as plt

# 1. Plot a Straight Line: y = 2x + 1
x = np.linspace(-10, 10, 100)
y = 2 * x + 1

plt.figure()
plt.plot(x, y)
plt.title("Straight Line: y = 2x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

# 2. Plot a Sine Wave
x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

plt.figure()
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()

# 3. Bar Chart of Student Marks
students = ["A", "B", "C", "D", "E"]
marks = [85, 92, 78, 88, 95]

plt.figure()
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()

# 4. Scatter Plot (Height vs Weight)
heights = [150, 155, 160, 165, 170, 175, 180]   # in cm
weights = [50, 54, 59, 62, 68, 72, 78]          # in kg

plt.figure()
plt.scatter(heights, weights)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True)
plt.show()

# 5. Show Category Distribution (e.g., count of students in grades)
grades = ["A", "B", "A", "C", "B", "A", "D", "B", "C", "A"]
unique_grades = sorted(set(grades))
counts = [grades.count(g) for g in unique_grades]

plt.figure()
plt.bar(unique_grades, counts)
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Count")
plt.show()

# 6. Histogram of Random Data
data = np.random.randn(1000)  # 1000 random numbers from normal distribution

plt.figure()
plt.hist(data, bins=30)
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Q6: Load Iris.csv and perform basic analysis with pandas

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("Iris.csv")

# View first few rows
print("First 5 rows of the dataset:")
print(df.head(), "\n")

# Check the shape of the dataset
print("Shape of the dataset (rows, columns):")
print(df.shape, "\n")

# Summarize numerical features
print("Statistical summary of numerical features:")
print(df.describe(), "\n")

# Identify feature names
print("Column names (feature names):")
print(df.columns.tolist(), "\n")

# Count samples in each class
target_col = "Species"
print(f"Number of samples in each class (column: {target_col}):")
print(df[target_col].value_counts(), "\n")

# Find missing values
print("Missing values in each column:")
print(df.isnull().sum(), "\n")

# Select only one column (Sepal Length)
one_col = "SepalLengthCm"
print(f"Values from column '{one_col}':")
print(df[one_col].head(), "\n")

# Select multiple columns
multi_cols = ["SepalLengthCm", "SepalWidthCm"]
print(f"Values from columns {multi_cols}:")
print(df[multi_cols].head(), "\n")

# Filter rows (conditional selection): Sepal Length > 5.0
filtered = df[df["SepalLengthCm"] > 5.0]
print("Rows where SepalLengthCm > 5.0:")
print(filtered.head(), "\n")

# Correlation matrix (numerical columns)
print("Correlation matrix of numerical features:")
print(df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]].corr(), "\n")

# Basic plot from pandas: scatter plot Sepal Length vs Petal Length
df.plot(
    kind="scatter",
    x="SepalLengthCm",
    y="PetalLengthCm",
    title="Sepal Length vs Petal Length"
)
plt.show()
```
